# Palettes
Palettes are not built in, however, they are useful for switching between colors and will be used in the rest of the examples

## Palette Definitions
Here, the different palettes are defined. Each Color takes in 4 values between 0-255 in the order of: Red, Green, Blue, Aplha(Opacitity/Transparency)

```python
from hex_renderer_py import Color

palettes = {
  "default": [
      Color(255, 107, 255, 255),
      Color(168, 30, 227, 255),
      Color(100, 144, 237, 255),
      Color(177, 137, 199, 255),
  ],
  "turbo": [
    Color(63, 61, 156, 255),
    Color(64, 150, 254, 255),
    Color(25, 227, 184, 255),
    Color(132, 254, 80, 255),
    Color(223, 222, 54, 255),
    Color(253, 140, 39, 255),
    Color(214, 52, 5, 255),
    Color(122, 4, 2, 255),
  ],
  "dark2": [
    Color(27, 158, 119, 255),
    Color(217, 95, 2, 255),
    Color(117, 112, 179, 255),
    Color(231, 41, 138, 255),
    Color(102, 166, 30, 255),
    Color(230, 171, 2, 255),
    Color(166, 118, 29, 255),
    Color(102, 102, 102, 255),
  ],
  "tab10": [
    Color(31, 119, 180, 255),
    Color(255, 127, 14, 255),
    Color(44, 160, 44, 255),
    Color(214, 39, 40, 255),
    Color(148, 103, 189, 255),
    Color(140, 86, 75, 255),
    Color(227, 119, 194, 255),
    Color(127, 127, 127, 255),
    Color(188, 189, 34, 255),
    Color(23, 190, 207, 255),
  ]
}
```

## Examples
Below are some examples of a pattern being rendered with all 4 palettes that will be used for the examples using a simple segment renderer.

```python
from hex_renderer_py import PatternVariant, HexGrid, GridOptions, Point, GridPatternOptions, Intersections, Lines, Triangle, CollisionOption
from IPython.display import display, Image

for palette in palettes:
    segment = GridOptions(
        line_thickness=0.12,
        center_dot=Point.None_(),
        pattern_options=GridPatternOptions.Uniform(
            intersections=Intersections.Nothing(),
            lines=Lines.SegmentColors(
                colors=palettes[palette], # <- palette change
                triangles=Triangle.Match(0.2),
                collisions=CollisionOption.Dashes(Color(255,0,0,255))
            )
        ),
    )
    pattern = [PatternVariant("NORTH_WEST", "wawqwawwwewwwewwwawqwawwwewwwewdeaweewaqaweewaawwww")]
    hex_grid = HexGrid(pattern, 50)

    img = hex_grid.draw_png(50, segment)

    print(palette)
    display(Image(data=bytes(img)))
```
