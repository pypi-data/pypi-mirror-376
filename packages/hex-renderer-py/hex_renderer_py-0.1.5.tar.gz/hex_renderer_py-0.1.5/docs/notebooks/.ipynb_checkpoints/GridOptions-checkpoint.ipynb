## GridOptions
GridOptions is the start of all pattern rendering settings and has 3 fields: line_thickness, center_dot, and pattern_options. 
* Line thickness is a bit self-explanatory, but it's simply how thick to draw the lines and is measured as a percentage of the line length. So, the actual line_thickness will be line_thickness * line_length.
* Center dot is the setting for (optionally) drawing a dot in the center of certain patterns to make the size of a pattern clearer.
* Pattern options are where the rest of the rendering settings lie and determines some extra behavior for changing them when seeing a pattern like introspection or retrospection

### Line Thickness

```python
from hex_renderer_py import GridOptions, GridPatternOptions, Point, Lines, Intersections, HexGrid

for line_thickness in range(12, 31, 4):
    line_thickness = line_thickness/100
    
    monocolor = GridOptions(
        line_thickness=line_thickness,
        pattern_options=GridPatternOptions.Uniform(
            intersections=Intersections.Nothing(),
            lines=Lines.Monocolor(
                color=palettes["default"][0],
                bent= True
            )
        ),
        center_dot=Point.None_()
    )

    grid = HexGrid(example_patterns, 50)

    print("Line Thickness:", line_thickness)
    display(Image(data=bytes(grid.draw_png(25, monocolor))))
```

### Center Dot

```python
from hex_renderer_py import GridOptions, GridPatternOptions, Point, Lines, Intersections, HexGrid, Marker, Color

    
monocolor = GridOptions(
    line_thickness=0.12,
    pattern_options=GridPatternOptions.Uniform(
        intersections=Intersections.Nothing(),
        lines=Lines.Monocolor(
            color=palettes["default"][0],
            bent= True
        )
    ),
    center_dot=Point.None_()
)

patterns = [
    PatternVariant("WEST", "qqq"),
    PatternVariant("NORTH_EAST", "eeeee"),
    PatternVariant("NORTH_EAST", "wewewewewew"),
    PatternVariant("EAST", "eee"),
    PatternVariant("NORTH_EAST", "qaq")
]
grid = HexGrid(patterns, 50)

print("No Center Dot")
display(Image(data=bytes(grid.draw_png(25, monocolor))))

monocolor = monocolor.with_center_dot(
    Point.Single(
        Marker(
            color=Color(255,0,0,255),
            radius=0.14,
        )
    )
)
print("Center Dot")
display(Image(data=bytes(grid.draw_png(25, monocolor))))
```
